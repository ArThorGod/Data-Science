```
import pandas as pd
# força o pandas não omitir nenhuma coluna
pd.set_option('display.max_columns', 21)

file_path = '/content/drive/MyDrive/Unifeb/2021/Aulas/IA/algoritmos/datasets/'
file_name = 'kc_house_data.csv'

file = pd.read_csv(file_path + file_name)

file.head()
```

```
# excluindo features irrelevantes
# O parâmetro axis = 1 que significa irá excluir uma coluna e não uma linha
# O parâmetro inplace é usado para persistir a alteração na variável dataset
# em memória, sem ele, a alteração não seria efetivada no objeto dataset.
file.drop('id', axis = 1, inplace = True)
file.drop('date', axis = 1, inplace = True)
file.drop('zipcode', axis = 1, inplace = True)
file.drop('lat', axis = 1, inplace = True)
file.drop('long', axis = 1, inplace = True)

file.head()
```

```
# definindo variáveis preditoras e variável algo
# Ou variáveis Independentes X e dependente Y
y = file['price']
x = file.drop('price', axis = 1)
```

```
from sklearn.model_selection import train_test_split
# funcão que reparte os dados em dados de treino e dados de teste
# test_size = 0.3 tamanho dos dados de teste = 30%
# random_state = garante a mesma aleatoriedade para todos
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 10)
```

```
from sklearn.linear_model import LinearRegression
# Criação do modelo de Machine Learning
model = LinearRegression()
# Alimentando o modelo com os dados Independente 'Target' e Dependentes 'Target'
model.fit(x_train, y_train)
```

```
# Coeficiente de Determinação R2
# O R-quadrado é uma medida estatística de quão próximos os dados estão da linha de regressão ajustada
# Ele também é conhecido como o coeficiente de determinação ou o coeficiente de determinação múltipla para a regressão múltipla.
result = model.score(x_test, y_test)
print(result)
# O resultado também pode ser chamado de acurácia (precisão), que no caso foi de 90%
```
