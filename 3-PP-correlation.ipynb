```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

file_path = '/content/drive/MyDrive/Unifeb/2021/Aulas/IA/algoritmos/datasets/'
file_name = 'diabetes.csv'

columns = ['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age','Outcome']

file = pd.read_csv( file_path + file_name, names = columns, header = 0 )

file.head()
```

```
# mostra a correlação
file.corr( method = 'pearson' )
```

```
# configurando uma figura de tamanho 10 x 10
plt.figure( figsize = (10, 10) )

# set heatmap setup
sb.heatmap( file.corr( method = 'pearson' ) )

# show image
plt.show()
```
