```
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
# gerando uma massa de dados
# n_samples = quantidade de dados
# n_features = quantidade de variáveis independentes
# noise = ruido quanto maior mais difícil ter uma correlação direta a uma reta
x, y = make_regression(n_samples=10, n_features=1, noise=10)
# print(x)
# print(y)

# acrescenta os dados para serem apresentados no gráfico
plt.scatter(x, y)
# mostrando os dados no gráfico
plt.show()
```

```
from sklearn.linear_model import LinearRegression
# Criação do modelo de Machine Learning
model = LinearRegression()
# Alimentando o modelo com os dados Independente 'Target' e Dependentes 'Target'
model.fit(x, y)
```

```
# Verificando os Resultados
# Coeficiente Linear - Onde a reta irá cortar nos pontos X e Y
print(model.intercept_)
# Coeficiente Angular - Grau de inclinação da reta a partir da variável Target
# retorna um vetor pois há a possibilidade de ter inúmeras variáveis Independentes
print(model.coef_)
```

```
import numpy as np
plt.scatter(x, y)
# Cria uma definição a partir de -3 até 3 indo de 1 em 1
xreg = np.arange(-3, 3, 1)
# 47.24 = Coeficiente Angular
# 0.6457 = Coeficiente Linear
# vem da expressão y = mx + b (Função Linear - Equação de Retas)
plt.plot(xreg, 93.70 * xreg - 10.44, color='red')
plt.show()
```

```
from sklearn.model_selection import train_test_split
# funcão que reparte os dados em dados de treino e dados de teste
# test_size = 0.3 tamanho dos dados de teste = 30%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3)
# alimentando o modelo com os dados de treino
model.fit(x_train, y_train)
```

```
# Coeficiente de Determinação R2
# O R-quadrado é uma medida estatística de quão próximos os dados estão da linha de regressão ajustada
# Ele também é conhecido como o coeficiente de determinação ou o coeficiente de determinação múltipla para a regressão múltipla.
result = model.score(x_test, y_test)
print(result)
# O resultado também pode ser chamado de acurácia (precisão), que no caso foi de 90%
```
